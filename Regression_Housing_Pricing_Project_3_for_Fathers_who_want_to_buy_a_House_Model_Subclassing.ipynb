```
import tensorflow as tf
import pandas
import keras
from keras.utils import to_categorical
from keras.layers import Dense
```

```
# load dataset
dataframe = pandas.read_csv("https://raw.githubusercontent.com/jbrownlee/Datasets/master/housing.data", delim_whitespace=True, header=None)
dataset = dataframe.values
# split into input (X) and output (Y) variables
X = dataset[:,0:13]
Y = dataset[:,13]
```

```
#Baseline
class ML(keras.Model):

  def __init__(self):
    super(ML, self).__init__()
    inputs = keras.Input(shape=[(13)])
    self.L1 = Dense(7, activation='relu')
    self.L2 = Dense(1, activation='softmax')

  def call(self, inputs):
    x = self.L1(inputs)
    return self.L2(x)
```

```
model = ML()
model.compile(optimizer='Adam', loss='mean_squared_error', metrics=['mae'])
model.fit(X,Y,epochs=500)
```

```
#smaller Network
class ML(keras.Model):
  def __init__(self):
    super(ML,self).__init__()
    inputs = keras.Input(shape=[(13)])
    self.L1 = Dense(3, activation='relu')
    self.L2 = Dense(1, activation='softmax')

  def call(self,inputs):
    x = self.L1(inputs)
    return self.L2(x)
```

```
model = ML()
model.compile(optimizer='Adam', loss='mean_squared_error', metrics=['mae'])
model.fit(X,Y, epochs=100)
```

```
#create larger Network
class ML(keras.Model):

  def __init__(self):
    super(ML,self).__init__()
    inputs = keras.Input(shape=[(13)])
    self.L1 = Dense(10,activation='relu')
    self.L2 = Dense(3,activation='softmax')

  def call(self, inputs):
    x=self.L1(inputs)
    return self.L2(x)
```

```
model=ML()
model.compile(optimizer='Adam', loss='mean_squared_error', metrics=['mae'])
model.fit(X,Y, epochs=100)
```

```
#Overfitting

class ML(keras.Model):
  def __init__(self):
    super(ML,self).__init__()
    inputs = keras.Input(shape=[(13)])
    self.L1 = Dense(512, activation='relu')
    self.L2 = Dense(128, activation='relu')
    self.L3 = Dense(64, activation='relu')
    self.L4 = Dense(32, activation='relu')
    self.L5 = Dense(1, activation='softmax')

  def call(self,inputs):
    x = self.L1(inputs)
    x = self.L2(x)
    x = self.L3(x)
    x = self.L4(x)
    return self.L5(x)
```

```
model=ML()
model.compile(optimizer='Adam', loss='mean_squared_error', metrics=['mae'])
model.fit(X,Y, epochs=2000)
```
