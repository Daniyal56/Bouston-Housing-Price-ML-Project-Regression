```
import tensorflow as tf
import pandas
import keras
from keras.utils import to_categorical
from keras.layers import Dense
```

```
 #load dataset
dataframe = pandas.read_csv("https://raw.githubusercontent.com/jbrownlee/Datasets/master/housing.data", delim_whitespace=True, header=None)
dataset = dataframe.values
# split into input (X) and output (Y) variables
X = dataset[:,0:13]
Y = dataset[:,13]
```

```
inputs = keras.Input(shape=[(13)])
x = Dense(7, activation='relu')(inputs)
outputs = Dense(1,activation='linear')(x)
```

```
model = keras.Model(inputs,outputs)
model.compile(optimizer='Adam', loss='mean_squared_error', metrics=['mae'])
```

```
model.fit(X,Y, epochs=500)
```

```
#Step 5. Tuning Layers and Number of Neurons in The Model 
#Evaluate a Smaller Network

inputs = keras.Input(shape=(13,))
x = Dense(3, activation='relu')(inputs)
outputs = Dense(1, activation='linear')(x)

model = keras.Model(inputs,outputs)
model.compile(optimizer='Adam', loss='mean_squared_error', metrics=['mae'])
model.fit(X,Y,epochs=100)
```

```
#Evaluate a Larger Network

inputs= keras.Input(shape=[(13)])
x = Dense(10, activation='relu')(inputs)
outputs = Dense(1,activation='linear')(x)

#Model

model = keras.Model(inputs,outputs)
model.compile(optimizer='Adam', loss = 'mean_squared_error', metrics=['mae'])
model.fit(X,Y,epochs=100)
```

```
#Overfitting

inputs = keras.Input(shape=(13,))
x = Dense(512, activation='relu')(inputs)
x = Dense(256, activation='relu')(x)
x = Dense(128, activation='relu')(x)
x = Dense(64, activation='relu')(x)
x = Dense(32, activation='relu')(x)
outputs = Dense(1, activation='linear')(x)

#model
model = keras.Model(inputs,outputs)
model.compile(optimizer='Adam', loss='mean_squared_error', metrics=['mae'])
model.fit(X,Y,epochs=1000)
```

```
#Tunning

inputs= keras.Input(shape=(13,))
x = Dense(8, activation='relu')(inputs)
outputs = Dense(1,activation='linear')(x)

#Model

model = keras.Model(inputs,outputs)
model.compile(optimizer='Adam', loss = 'mean_squared_error', metrics=['mae'])
model.fit(X,Y,epochs=70)
 
```
