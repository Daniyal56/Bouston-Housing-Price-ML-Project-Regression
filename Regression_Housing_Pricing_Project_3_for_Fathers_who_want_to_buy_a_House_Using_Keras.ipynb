```python
import numpy
import pandas
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import tensorflow as tf
```

```python
# load dataset
dataframe = pandas.read_csv("https://raw.githubusercontent.com/jbrownlee/Datasets/master/housing.data", delim_whitespace=True, header=None)
dataset = dataframe.values
# split into input (X) and output (Y) variables
X = dataset[:,0:13]
Y = dataset[:,13]
```

```python
model = Sequential()
model.add(Dense(7, activation='relu', input_shape=[(13)]))
model.add(Dense(1, activation='linear'))
```

```python
#creating model

model = Sequential()
model.add(Dense(7, activation='relu', input_shape=[(13)]))
model.add(Dense(1, activation='linear'))

model.compile(optimizer='Adam', loss='mean_squared_error', metrics=['mae'])
model.fit(X,Y,epochs=200)

```

```python
#smaller Network

model = Sequential()
model.add(Dense(4, activation='relu', input_shape=[(13)]))
model.add(Dense(1, activation='linear'))

model.compile(optimizer='Adam', loss='mean_squared_error', metrics=['mae'])
model.fit(X,Y,epochs=200)

```

```python
#larger Network

model = Sequential()
model.add(Dense(10, activation='relu', input_shape=[(13)]))
model.add(Dense(1, activation='linear'))

model.compile(optimizer='Adam', loss='mean_squared_error', metrics=['mae'])
model.fit(X,Y,epochs=200)

```

```python
#Overfitting Network

model = Sequential()
model.add(Dense(512, activation='relu', input_shape=[(13)]))
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='linear'))

model.compile(optimizer='Adam', loss='mean_squared_error', metrics=['mae'])
model.fit(X,Y,epochs=1000)

```
